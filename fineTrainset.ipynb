```python
import pandas as pd
df = pd.read_excel('./데이터정제/df_train.xlsx')
```

```python
df
```

```python
df.to_csv('df_train.csv')
```

```python
df_train1 = pd.DataFrame(columns=['phrase', 'aspect', 'score'])
df_train2 = pd.DataFrame(columns=['phrase', 'aspect', 'score'])
df_train3 = pd.DataFrame(columns=['phrase', 'aspect', 'score'])
df_train4 = pd.DataFrame(columns=['phrase', 'aspect', 'score'])
df_train5 = pd.DataFrame(columns=['phrase', 'aspect', 'score'])
df_train6 = pd.DataFrame(columns=['phrase', 'aspect', 'score'])
```

```python
df_train1['phrase'] = df['diligence.1']
df_train2['phrase'] = df['leadership.1']
df_train3['phrase'] = df['responsibility.1']
df_train4['phrase'] = df['passion.1']
df_train5['phrase'] = df['conductivity/creativity.1']
df_train6['phrase'] = df['cooperation.1']

df_train1['aspect'] = 'diligence'
df_train2['aspect'] = 'leadership'
df_train3['aspect'] = 'responsibility'
df_train4['aspect'] = 'passion'
df_train5['aspect'] = "conductivity/creativity"
df_train6['aspect'] = "cooperation"

df_train1['score'] = df['diligence']
df_train2['score'] = df['leadership']
df_train3['score'] = df['responsibility']
df_train4['score'] = df['passion']
df_train5['score'] = df['conductivity/creativity']
df_train6['score'] = df['cooperation']
```

```python
df_train = pd.concat([df_train1, df_train2, df_train3, df_train4, df_train5, df_train6])
```

```python
df_train.dropna(subset=["phrase"], axis=0, inplace=True)
```

```python
df_train.to_excel('df_train.xlsx', index=False)
```
