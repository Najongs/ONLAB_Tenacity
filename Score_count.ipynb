```python
#  

works = pd.read_excel("/Users/najong-yeol/Library/CloudStorage/OneDrive-/ /NAJO/()/ //  .xlsx")
works = works.loc[:, ['', '']]
works = works.dropna(axis=0)
workname = list(works[''])
workscore = list(works[''])
work_class = dict(zip(workname, workscore))
work_score = {'A' : 1, 'B' : 0.9, 'C' : 0.8, 'D' : 0.7}
```

```python
work_class
```

```python
#   

#   
input_ch = {'userID':'00001', 'voteID':'00002', 'work_time(h)':30, 'work_name': '', 'checklist':[4,5,6], 'text':'  ... !'}

pas = [1,2,3]
coop = [4,5,6]
dil = [7,8,9]
res = [10,11,12]
con = [13,14,15]
lead = [16,17,18]

data['00001']['total_work_time'] += input_ch['work_time(h)']

for p in pas:
    if p in input_ch['checklist']:
        data['00001']['checklist']['passion_ch'] += 1*input_ch['work_time(h)']*work_score[work_class[input_ch['work_name']]]
        data['00001']['votePoint'] += 1
for p in coop:
    if p in input_ch['checklist']:
        data['00001']['checklist']['cooperation_ch'] += 1*input_ch['work_time(h)']*work_score[work_class[input_ch['work_name']]]
        data['00001']['votePoint'] += 1
for p in dil:
    if p in input_ch['checklist']:
        data['00001']['checklist']['diligence_ch'] += 1*input_ch['work_time(h)']*work_score[work_class[input_ch['work_name']]]
        data['00001']['votePoint'] += 1
for p in res:
    if p in input_ch['checklist']:
        data['00001']['checklist']['responsibility_ch'] += 1*input_ch['work_time(h)']*work_score[work_class[input_ch['work_name']]]
        data['00001']['votePoint'] += 1
for p in con:
    if p in input_ch['checklist']:
        data['00001']['checklist']['conductivity_ch'] += 1*input_ch['work_time(h)']*work_score[work_class[input_ch['work_name']]]
        data['00001']['votePoint'] += 1
for p in lead:
    if p in input_ch['checklist']:
        data['00001']['checklist']['leadership_ch'] += 1*input_ch['work_time(h)']*work_score[work_class[input_ch['work_name']]]
        data['00001']['votePoint'] += 1
        
data['00001']['voteCount'] += 1
```

```python
#   
model = 0

#  
model_output_tx = {'passion_tx' : 100, 'cooperation_tx':100, 'diligence_tx':100, 'responsibility_tx':100, 'conductivity_tx':100, 'leadership_tx':100 }

data['00001']['answerCount'] += 1

data['00001']['text_score']['passion_tx'] += model_output_tx['passion_tx']
data['00001']['text_score']['cooperation_tx'] += model_output_tx['cooperation_tx']
data['00001']['text_score']['diligence_tx'] += model_output_tx['diligence_tx']
data['00001']['text_score']['responsibility_tx'] += model_output_tx['responsibility_tx']
data['00001']['text_score']['conductivity_tx'] += model_output_tx['conductivity_tx']
data['00001']['text_score']['leadership_tx'] += model_output_tx['leadership_tx']
```
